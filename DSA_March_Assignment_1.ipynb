# Import numpy,pandas,matplotlib and seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the dataset in excel format to python environment

```python
data = pd.read_excel(r'C:\Users\USER\Downloads\iris.xls')
```

# Check for the null values present in the dataset

```python
# Finding missing values using isnull() function.If there is a data missing,it will display True else False

data.isnull()
```

# Perform visualizations using the dataset

```python
# Histogram of Sepal Length

plt.hist(data['SL'],bins=20,color='green')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Count')
plt.title('Sepal Length')
plt.show()
```

```python
# Histogram of Sepal Width

plt.hist(data['SW'],bins=20,color='red')
plt.xlabel('Sepal Width in cm')
plt.ylabel('Count')
plt.title('Sepal Width')
plt.show()
```

```python
# Countplot of Petal length

plt.figure(figsize=(18,8))
sns.countplot(x=data.PL)
plt.title('Petal Length')
plt.show()
```

```python
# Countplot of Sepal length

plt.figure(figsize=(18,8))
sns.countplot(x=data.SL)
plt.title('Sepal Length')
plt.show()
```

```python
# Boxplot of Sepal Width

plt.boxplot(data['SW'])
plt.xlabel('Sepal Width in cm')
plt.ylabel('Count')
plt.title('Sepal Width')
plt.show()
```

```python
# Boxplot of Sepal Length


plt.boxplot(data['SL'])
plt.xlabel('Sepal Length in cm')
plt.ylabel('Count')
plt.title('Sepal Length')
plt.show()
```

```python
# Boxplot of Petal Length


plt.boxplot(data['PL'])
plt.xlabel('Petal Length in cm')
plt.ylabel('Count')
plt.title('Petal Length')
plt.show()
```

```python
# Boxplot of Petal Width


plt.boxplot(data['PW'])
plt.xlabel('Petal Width in cm')
plt.ylabel('Count')
plt.title('Petal Width')
plt.show()
```

```python
# Scatter Plot  of sepal width and classification

plt.scatter(data['SW'],data['Classification'])
plt.xlabel('Sepal Width in cm')
plt.ylabel('Count')
plt.title('Sepal Width')
plt.show()
```

```python
# Scatter Plot of sepal length and classification

plt.scatter(data['SL'],data['Classification'])
plt.xlabel('Sepal Length in cm')
plt.ylabel('Count')
plt.title('Sepal Length')
plt.show()
```

```python
# Scatter plot of sepal length and sepal width

plt.scatter(data['SL'],data['SW'])
plt.xlabel('Sepal Length in cm')
plt.ylabel('Sepal Width in cm')
plt.title('Relation between Sepal Length & Sepal Width')
plt.show()
```
