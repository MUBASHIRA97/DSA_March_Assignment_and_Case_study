```python
# Handling outliers using Flooring And Capping Method:

Q1 = train['ApplicantIncome'].quantile(0.25)
Q3 = train['ApplicantIncome'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
train['ApplicantIncome']=np.where(train['ApplicantIncome']>upper_whisker,upper_whisker,
                               np.where(train['ApplicantIncome']<lower_whisker,lower_whisker,train['ApplicantIncome']))
```

# Encoding

```python
# Import label encoding library
from sklearn.preprocessing import LabelEncoder

# Function to apply label encoding
def encode_labels(data):
    for col in data.columns:
        if data[col].dtype == 'object':
            le = LabelEncoder()
            data[col] = le.fit_transform(data[col])
  
    return data
  
# Applying function in whole column of train data
train = encode_labels(train)
```

```python
# Import label encoding library
from sklearn.preprocessing import LabelEncoder

# Function to apply label encoding
def encode_labels(data):
    for col in data.columns:
        if data[col].dtype == 'object':
            le = LabelEncoder()
            data[col] = le.fit_transform(data[col])
  
    return data
  
# Applying function in whole column of test data
test = encode_labels(test)
```

# scaling

```python
# Import StandardScaler library
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

train['Loan_ID'] = scaler.fit_transform(train['Loan_ID'].values.reshape(-1, 1))
test['Loan_ID'] = scaler.transform(test['Loan_ID'].values.reshape(-1, 1))
                                   
numerical_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
 
for col in numerical_cols:
    train[col] = scaler.fit_transform(train[col].values.reshape(-1, 1))
    test[col] = scaler.transform(test[col].values.reshape(-1, 1))                                   
```

# splitting the data

```python
X = train.drop('Loan_Status', axis = 1)
y = train['Loan_Status']
X_test = test
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Building model

```python
# Logistic regression model

# Import logistic regression

from sklearn.linear_model import LogisticRegression

# Create instance for logistic regression

logit_model = LogisticRegression()

# Fitting the model

logit_model.fit(X_train,y_train)

# prediction

y_pred_lr = logit_model.predict(X_val)
```

```python
# Evaluation of Logistic regression model

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

print('Accuracy :', accuracy_score(y_val,y_pred_lr))
print('Precision :', precision_score(y_val,y_pred_lr,average = 'micro'))
print('Recall :', recall_score(y_val,y_pred_lr,average = 'micro'))
print('F1 Score :', f1_score(y_val,y_pred_lr,average =  'micro'))
```

```python
# KNN model

from sklearn.neighbors import KNeighborsClassifier
metric = []
neighbors = np.arange(3,15)

for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski', p=2)
    classifier.fit(X_train,y_train)
    y_pred_knn = classifier.predict(X_val)
    acc = accuracy_score(y_val,y_pred_knn)
    metric.append(acc)
```

```python
# Import kNN and SVM Model libraries

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# create an istant of kNN and SVM Models

knn = KNeighborsClassifier()
svm = SVC()
```

```python
# Import k-fold and stratified k-fold cross validation libraries 

from sklearn.model_selection import cross_val_score,StratifiedKFold,KFold

# Define the number of fold

k =5

# k-fold cross validation

k_fold_scores_knn = cross_val_score(knn,X,y,cv=KFold(n_splits=k))
k_fold_scores_svm = cross_val_score(svm,X,y,cv=KFold(n_splits=k))

# Stratified k-fold cross validation

stratified_k_fold_scores_knn = cross_val_score(knn,X,y,cv=StratifiedKFold(n_splits=k))
stratified_k_fold_scores_svm = cross_val_score(svm,X,y,cv=StratifiedKFold(n_splits=k))
```

```python
# calculate the average accuracy scores

average_accuracy_knn = k_fold_scores_knn.mean()
average_accuracy_svm = k_fold_scores_knn.mean()
average_accuracy_stratified_knn = stratified_k_fold_scores_knn.mean()
average_accuracy_stratified_svm = stratified_k_fold_scores_svm.mean()

print('Average accuracy of k-fold kNN :',average_accuracy_knn)
print('Average accuracy of k-fold SVM :',average_accuracy_knn)
print('Average accuracy of Stratified k-fold kNN :',average_accuracy_stratified_knn)
print('Average accuracy of Stratified k-fold SVM :',average_accuracy_stratified_svm)
```

Here I'm selecting logistic regression,knn and svm model for hyperparameter tuning basedon accuracy score

```python
params = {
    'LogisticRegression' : {
        'model' : LogisticRegression(),
        'param' : {
            'penalty' : ['l2'], 
            'C' : [0.1, 0.3, 0.5, 0.7, 0.9, 1.0],
            'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
            'max_iter' : [100, 1000,2500, 5000]
        }
    },
    'SVC' : {
        'model' : SVC(),
        'param' : {
            'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
            'C' : [0.1, 0.3, 0.5, 0.7, 0.9, 1.0],
            'degree' : [3, 5]
        }
    },
    'KNN' : {
        'model' : KNeighborsClassifier(),
        'param' : {
               'n_neighbors' : [5,7,9,11,13,15],
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan']
        }
    }
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

scores = []

for name, mp in params.items():
    rs = RandomizedSearchCV(estimator=mp['model'], param_distributions=mp['param'], cv = 10,
                            verbose=3, n_jobs=-1)
    rs.fit(X_train, y_train)
    scores.append({
        'model' : name,
        'score' : rs.best_score_,
        'estimator' : rs.best_estimator_
    })
```

```python
final = pd.DataFrame(scores, columns=['model', 'score', 'estimator'])
final.sort_values('score', ascending=False)
```
