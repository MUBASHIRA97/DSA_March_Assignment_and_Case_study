```python
# Import numpy,pandas,matplotlib and seaborn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the datset

data = pd.read_csv(r'C:\Users\USER\Downloads\hotel-tips.csv')
```

```python
# Display columns in dataset

data.columns
```

```python
# Display information in dataset

data.info()
```

```python
# Display first 5 rows

data.head()
```

```python
# Weekdays present in dataset

data['day'].unique()
```

```python
# No. of dinners and lunches present

data['time'].value_counts()
```

```python
# Bar plot representation of no. of dinners and lunches

data['time'].value_counts().plot.bar()
plt.xlabel('Time')
plt.ylabel('Count')
plt.show()
```

This means more dinner is present in the given dataset

```python
# Percentage of dinner and lunch

data['time'].value_counts(normalize=True)
```

```python
# Which sex pay more tip

data.groupby(['sex'])['tip'].mean()
```

```python
# Bar plot representation of which sex more tip

data.groupby(['sex'])['tip'].mean().plot.bar(color='blue')
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.show()
```

This means male gives more tip

```python
# Smoker or non smoker who give more tip

data.groupby(['smoker'])['tip'].mean()
```

```python
# Bar plot representation of Smoker or non smoker who give more tip

data.groupby(['smoker'])['tip'].mean().plot.bar(color='red')
plt.xlabel('Smoker')
plt.ylabel('Tip')
plt.show()
```

```python
# Which sex smoker or non smoker give more tip

data.groupby(['smoker','sex'])['tip'].mean().plot.bar(color='g')
plt.show()
```

non smoker male gives more tip

```python
# Which sex smoker or non smoker give more tip

data.groupby(['smoker','sex'])['tip'].mean()
```

```python
# During lunch or dinner does the people pay more tip

data.groupby(['time'])['tip'].mean()
```

```python
# Bar plot representation of During lunch or dinner does the people pay more tip

data.groupby(['time'])['tip'].mean().plot.bar(color='indigo')
plt.xlabel('time')
plt.ylabel('tip')
plt.show()
```

People pay more tips during dinner

```python
# Bar plot of total bill generated each day

plt.bar(data['day'],data['total_bill'],color='green')
plt.xlabel('Day')
plt.ylabel('Total_bill')
plt.title('Total bill generated each day')
plt.show()
```

More bills are generated during saturday and less bills during friday

```python
# Bar plot of total bill generated by male and female

plt.bar(data['sex'],data['total_bill'],color='red')
plt.xlabel('Sex')
plt.ylabel('Total_bill')
plt.title('Total bill generated by male and female')
plt.show()
```

Males contributes more to the total bill

```python
# Bar plot of total bill generated by size

plt.bar(data['size'],data['total_bill'],color='violet')
plt.xlabel('Size')
plt.ylabel('Total_bill')
plt.title('Total bill generated by size')
plt.show()
```

Size of 3 contributes more to the total bill and size of 1 contributes less

```python
# Bar plot of total bill generated by time

plt.bar(data['time'],data['total_bill'],color='yellow')
plt.xlabel('Time')
plt.ylabel('Total_bill')
plt.title('Total bill generated by time')
plt.show()
```

More total bill is generated during dinner

```python
# Histogram of total bill

plt.figure(figsize=(18,8))
plt.hist(data['total_bill'],bins=20,color='blue')
plt.xlabel('Total_bill')
plt.ylabel('Count')
plt.title('Total bill generated')
plt.show()
```

Maximum bill is generated between 10 and 20

```python
# Histogram of size

plt.hist(data['size'],bins=20,color='red')
plt.xlabel('Size')
plt.ylabel('Count')
plt.title('Size')
plt.show()
```

Most of the customers are size of 2

```python
# Histogram of tip

plt.hist(data['tip'],bins=20,color='green')
plt.xlabel('Tip')
plt.ylabel('Count')
plt.title('Tip')
plt.show()
```

More tip given is in between 2-3

```python
# Relationship between total bill and tip

plt.scatter(data['total_bill'],data['tip'])
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title('Relation between total bill and tip')
plt.show()
```

Most of the total bill is in the range of 10-20 and most of the tip given is in the range of 2-4. As total bill increases tip also increases but there is also some outliners

```python
# Relationship between total bill and tip in correlation with sex using seaborn library

sns.scatterplot(x='total_bill',y='tip',data=data,hue='sex')
plt.title('Relationship between total bill and tip in correlation with sex')
plt.show()
```

Male pay more total bill than female and females are concentrated to certain range of total bill

```python
# Relationship between total bill and tip in correlation with time using seaborn library

sns.scatterplot(x='total_bill',y='tip',data=data,hue='time')
plt.title('Relationship between total bill and tip in correlation with time')
plt.show()
```

Most of the total bill of dinner time concentrates in a certain range of 10-20 and most of the tip ranges between 2-4. Dinner  time pays the higher total bill and tip

```python
# Relationship between total bill and tip in correlation with size using seaborn library

sns.scatterplot(x='total_bill',y='tip',data=data,hue='size')
plt.title('Relationship between total bill and tip in correlation with size')
plt.show()
```

Most of the total bill is generated by size of 2. Maximum  total bill and tip was by size of 3
