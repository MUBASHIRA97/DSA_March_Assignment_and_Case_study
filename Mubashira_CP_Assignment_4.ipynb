# 5. Check and handle outliers in at least 3 columns in the dataset


Outliers can be identified by different methods:Visualization,Skewness,Interquartile Range,Standard Deviation

Skewness method : The skewness value should be within the range of -1 to 1 for a normal distribution,any major changes from 
this value may indicate the presence of outliers.

```python
print('skewness value of Age: ',df['Age'].skew())
print('skewness value of Fare: ',df['Fare'].skew())
print('skewness value of Pclass: ',df['Pclass'].skew())
print('skewness value of Survived: ',df['Survived'].skew())
```

From the skewness values of Age,Fare,Pclass and Survived , the skewness value of fare is out of range -1 to 1, this indicates the presence of outliers in Fare column

Outliers Treatment can be done by Flooring and Capping,Trimming and Replacing outliers with the mean, median, mode, or 
other value

```python
# Boxplot before Flooring And Capping Method: 

plt.boxplot(df['Fare'])
plt.show()
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
df['Fare']=np.where(df['Fare']>upper_whisker,upper_whisker,np.where(df['Fare']<lower_whisker,lower_whisker,df['Fare']))
```

```python
# Boxplot after Flooring And Capping Method: 

plt.boxplot(df['Fare'])
plt.show()
```

```python
print('skewness value of Fare: ',df['Fare'].skew())
```

This means no outliers
